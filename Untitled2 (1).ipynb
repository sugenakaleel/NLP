```python
import re
import pandas as pd
import nltk
import spacy
import matplotlib.pyplot as plt
import seaborn as sns

from nltk.corpus import stopwords
from textblob import TextBlob
```

```python
nlp = spacy.load("en_core_web_sm")
stop_words = set(stopwords.words("english"))
```

```python
df = pd.read_csv("customer_support_tickets.csv")
df.head()
```

```python
df.columns = df.columns.str.lower()
df = df.rename(columns={
    "ticket description": "raw_text",
    "ticket type": "category"
})

df = df[["raw_text", "category"]]
df.head()
```

```python
print("Dataset Shape:", df.shape)
print("\nMissing Values:\n")
print(df.isnull().sum())
```

```python
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
def preprocess_text(text):
    cleaned = clean_text(text)
    doc = nlp(cleaned)
    
    tokens = [
        token.lemma_
        for token in doc
        if token.text not in stop_words and not token.is_punct
    ]
    
    sentiment = TextBlob(cleaned).sentiment.polarity
    return " ".join(tokens), sentiment
```

```python
# Take a small representative sample (VERY IMPORTANT)
df_sample = df.sample(300, random_state=42).reset_index(drop=True)

df_sample.shape
```

```python
import spacy

# Disable slow components
nlp = spacy.load(
    "en_core_web_sm",
    disable=["ner", "parser", "textcat"]
)
```

```python
def fast_preprocess(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"[^a-zA-Z\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    
    doc = nlp(text)
    return " ".join(
        token.lemma_
        for token in doc
        if token.text not in stop_words and not token.is_punct
    )
```

```python
df_sample["processed_text"] = df_sample["raw_text"].apply(fast_preprocess)
df_sample.head()
```

```python
df_sample["sentiment"] = df_sample["raw_text"].apply(
    lambda x: TextBlob(str(x)).sentiment.polarity
)

df_sample.head()
```

```python
nlp_ner = spacy.load("en_core_web_sm")

def extract_entities(text):
    doc = nlp_ner(str(text))
    return [(ent.text, ent.label_) for ent in doc.ents]
```

```python
df_sample["entities"] = None

df_sample.loc[:10, "entities"] = df_sample.loc[:10, "raw_text"].apply(extract_entities)

df_sample.head(12)
```

```python
plt.figure(figsize=(8,4))
sns.countplot(x="category", data=df)
plt.title("Customer Support Ticket Category Distribution")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df_sample["sentiment"], bins=15)
plt.title("Sentiment Polarity Distribution")
plt.xlabel("Sentiment Score")
plt.show()
```

```python
sample_text = df.iloc[0]["raw_text"]

processed, sentiment = preprocess_text(sample_text)
entities = extract_entities(sample_text)

print("ORIGINAL TICKET:\n", sample_text)
print("\nPROCESSED TEXT:\n", processed)
print("\nSENTIMENT SCORE:\n", sentiment)
print("\nNAMED ENTITIES:\n", entities)
```
